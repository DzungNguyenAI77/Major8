```python
cache_dir = r'C:/ChuyenNganh_8/DSP_F1/f1-predictor/cache_dir'

if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

fastf1.Cache.enable_cache(cache_dir)

all_results = pd.DataFrame()

for round_num in range(1, 23):
    try:
        session = fastf1.get_session(2023, round_num, 'Q')
        session.load()

        results = session.results.loc[:, [
            'Abbreviation', 'TeamName', 'FullName', 'Q1', 'Q2', 'Q3'
        ]]

        results['Round'] = round_num

        all_results = pd.concat([all_results, results], ignore_index=True)

    except Exception as e:
        print(f"Failed to load session for round {round_num}: {e}")

all_results['Q1'] = all_results['Q1'].apply(lambda x: str(x)[-12:] if pd.notna(x) else 'NaT')
all_results['Q2'] = all_results['Q2'].apply(lambda x: str(x)[-12:] if pd.notna(x) else 'NaT')
all_results['Q3'] = all_results['Q3'].apply(lambda x: str(x)[-12:] if pd.notna(x) else 'NaT')

print(all_results)

csv_path = 'C:/ChuyenNganh_8/DSP_F1/f1-predictor/data_qual/f1_2023_qualifying_results.csv'
all_results.to_csv(csv_path, index=False)

csv_path 
```

```python
qual_data = pd.read_csv('C:/ChuyenNganh_8/DSP_F1/f1-predictor/data_qual/f1_2023_qualifying_results.csv')
qual_data.head(20)
```

```python
qual_data['qualifying_time'] = qual_data[['Q1', 'Q2', 'Q3']].min(axis=1)
```

```python
qual_data.drop(columns={'Q1', 'Q2', 'Q3'}, inplace=True)
qual_data
```

```python
qual_data['driver_name'] = qual_data['FullName'] + ' ' + qual_data['Abbreviation']
qual_data.drop(columns={'Abbreviation', 'FullName'}, inplace=True)
qual_data
```

```python
qual_data.rename(columns = {'TeamName':'car'}, inplace=True)
qual_data
```

```python
qual_data['grid_position'] = qual_data.groupby('Round')['qualifying_time'].rank(method='first')

qual_data['grid_position'] = qual_data['grid_position'].astype('int64')

print(qual_data[['grid_position', 'driver_name', 'car', 'qualifying_time', 'Round']])
```
